```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
url="https://stats.espncricinfo.com/ci/engine/records/bowling/most_wickets_career.html?class=2;id=6;type=team"
```

```python
headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'}
```

```python
page=requests.get(url,headers=headers)
page
```

```python
sour=BeautifulSoup(page.content)
```

```python
table=sour.find("table")
table
```

```python
data=table.find_all("tr")
data
```

```python
info=[]
for i in data:
    for j in i.find_all("td"):
        info.append(j.text)
info
```

```python
# Create dataframe

cricketinfo=pd.DataFrame({})
cricketinfo["Name"]=names
cricketinfo["Span"]=span
cricketinfo["Matches"]=match
cricketinfo["Innings"]=inning
cricketinfo["Balls"]=ball
cricketinfo["Over"]=over
cricketinfo["Maiden"]=maiden
cricketinfo["Runs"]=run
cricketinfo["Wicket"]=wicket
cricketinfo["BBI"]=bbi
cricketinfo["Avreage"]=avreage
cricketinfo["Economy"]=economy
cricketinfo["Strike rate"]=strike_rate
cricketinfo["four wickets in an innings"]=four_wickets
cricketinfo["five wickets in an innings"]=five_wickets
```

## Drop  unwanted colume

```python
df = df.drop(columns=['Unnamed: 0'])
```

## 1.Finding the top 10 bowlers based on wickets

```python
top_10_bowlers=df[["Name","Wicket"]].sort_values("Wicket",ascending=False).head(10)
top_10_bowlers
```

```python
sns.barplot(data=top_10_bowlers,y="Name",x="Wicket",palette='viridis')
plt.title('Top 10 Players Based on Wickets ', fontsize=16)
plt.xlabel('Wicket taken', fontsize=14)
plt.ylabel('Player Name', fontsize=14)
plt.show()
```

## 2.top 10 players Who play more matches

```python
more_matches = df[["Name","Matches"]].sort_values("Matches",ascending=False).head(10)
more_matches
```

```python
sns.barplot(data=more_matches,y="Name",x="Matches",palette='viridis')
plt.title('Top 10 Players Based on Matches Played', fontsize=16)
plt.xlabel('Matches Played', fontsize=14)
plt.ylabel('Player Name', fontsize=14)
plt.show()
```

## 3. top 10 Players vs Strike Rate

```python
high_strike_rate_bowlers = df[df['Strike rate'] > 50][["Name","Strike rate"]].sort_values("Strike rate",ascending=False).head(10)
high_strike_rate_bowlers
```

```python
# Plotting pie chart
plt.figure(figsize=(10, 8))
plt.pie(high_strike_rate_bowlers['Strike rate'], labels=high_strike_rate_bowlers['Name'], autopct='%1.1f%%')
plt.title('Bowlers with Strike Rate Greater than 50', fontsize=16)
plt.show()
```

## 4.Player till Active 

```python
# Extract end year from the 'Span' column
df['End Year'] = df['Span'].str.split('-').str[1].astype(int)

# Determine if the player is still active
current_year = pd.Timestamp.now().year
df['Still Active'] = df['End Year'] >= current_year

# Filter and display still active players
still_active_players = df[df['Still Active']]

# Display the result
print("Players Still Active:")
print(still_active_players[['Name', 'Span', 'End Year']])
```

## 5.Find the top 10 bowlers based on runs conceded

```python
top_10_runs=df[["Name","Over","Runs"]].sort_values("Runs",ascending=False).head(10)
top_10_runs
```

```python
# Plotting the bar chart
#plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_runs, x='Runs', y='Name', palette='coolwarm',hue="Over")
plt.title('Top 10 Bowlers Based on Runs and overs', fontsize=16)
plt.xlabel('Runs Conceded', fontsize=14)
plt.ylabel('Bowler Name', fontsize=14)
plt.show()
```

## 6.Top 5 Bowlers with Best Economy Rate:

```python
top_5_eco=df[["Name","Economy"]].sort_values("Economy",ascending=False).head(5)
top_5_eco
```

```python
#plt.figure(figsize=(12, 6))
sns.barplot(data=top_5_eco, y="Economy", x="Name", palette="Reds_r")
plt.title("Top 10 Bowlers with the Highest Economy Rates")
plt.xlabel("Economy Rate")
plt.ylabel("Bowler Name")
plt.show()
```
